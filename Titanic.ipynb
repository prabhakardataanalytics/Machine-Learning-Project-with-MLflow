```python
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("titanic")
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['age'].mean()
```

```python
data['age'].fillna(29,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
titanic_num=data.select_dtypes(["int64","float64"])
titanic_cat=data.select_dtypes("object")
```

```python
titanic_num
```

```python
titanic_cat
```

```python
from sklearn.preprocessing import LabelEncoder

for col in titanic_cat:
    le=LabelEncoder()
    titanic_cat[col]=le.fit_transform(titanic_cat[col])
```

```python
titanic_cat
```

```python
titanic_num
```

```python
data=pd.concat([titanic_num,titanic_cat],axis=1)
```

```python
data
```

```python
x=data.drop('alive',axis=1)
y=data['alive']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
for col in x:
    ss=StandardScaler()  # Standard scaler = x-mean/s.d
    xtrain[col]=ss.fit_transform(xtrain[[col]]) #2bracket for 2d array otherwise we get error
    xtest[col]=ss.transform(xtest[[col]])
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain,ytrain)
```

```python
y_pred=knn.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,y_pred))
```
